```python
total_market_return = (ind_capweight*ind_return).sum(axis="columns")
```

```python
total_market_return.plot()
```

```python
total_market_index = erk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(10,5), title="Total Market CapWeighted Index 1926-2018")
```

```python
total_market_index["1980":].plot(figsize=(12,6))
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(12,6))
```

```python
tml_tr36rets = total_market_return.rolling(window=36).aggregate(erk.annualize_rets, periods_per_year=12)
tml_tr36rets.plot(figsize=(12,6), label="Tr36 mo Return", legend=True)
total_market_return.plot(label="Return", legend=True)
```

### Rolling Correlation - along with MultiIndexes and '.groupby'

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ind_return.head()
```

```python
ts_corr.index.names = ["date", "industry"]
ts_corr.tail()
```

```python
ind_tr36corr = ts_corr.groupby(level='date').apply(lambda cormat: cormat.values.mean())
```

```python
ind_tr36corr.plot()
```

```python
tml_tr36rets.plot(label="Tr36 Mo Rets", figsize=(12,6), legend=True)
ind_tr36corr.plot(label="Tr36 Mo Corr", legend=True, secondary_y=True)
```

```python
tml_tr36rets.corr(ind_tr36corr)
```
